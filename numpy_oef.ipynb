#### Methode `maak_nul` (/2)

Implementeer een methode `maak_nul(x, index)` die een numpy array `x` en een index `index` meekrijgt en een array retourneert waarbij alle elementen vanaf index `0` t.e.m. index `index-1` gelijk zijn aan nul. Alle andere elementen blijven ongewijzigd.  Je mag de originele array `x` niet aanpassen! Hieronder een voorbeeld:

```python
maak_nul(np.array([-1, -2, 3, 4, 5]), 2)  # array([0,0,3,4,5])
```

````python
import numpy as np
````

````python
def maak_nul(x, index):
    # lengte = len(x)
    x[:index] = 0
    return x
````

````python
print(maak_nul(np.array([-1, -2, 3, 4, 5]), 2))
````

Implementeer een methode `eenheidsvector(n, i)` die een numpy array van lengte `n` teruggeeft
bestaande uit nullen behalve het element op plaats `i` dat gelijk is aan 1.

```python
eenheidsvector(5, 3)  # array([0., 0., 0., 1., 0.])

````python
def eenheidsvector(n, i):
    a = np.zeros(n)
    a[i] = 1
    return a
````

#### Methode `maak_spiegeling` (/3)

Schrijf een methode `maak_spiegeling(u)` die een numpy array `u` meekrijgt en een matrix retourneert die een Householder reflectie voorstelt.

De Householder reflectie is gedefinieerd als:
$$
\mathbf{H} = \mathbf{I} - 2\mathbf{v}\mathbf{v}^T,\quad \text{met } \mathbf{v} = \frac{\mathbf{u}}{\lVert \mathbf{u}\rVert}.
$$
Hierbij is $\mathbf{I}$ de eenheidsmatrix van de gepaste grootte en $\mathbf{v}\mathbf{v}^T$ is het **uitwendig product** van $\mathbf{v}$ met zichzelf.

Je kan wiskundig controleren dat $\mathbf{H}$ een orthogonale matrix is (maar je moet dit niet doen in deze opgave).

````python
def maak_spiegeling(u):
    v = u / np.linalg.norm(u)
    I = np.eye(len(u))
    return I - 2 * np.outer(v, v.T)
````

#### Implementatie van de QR-decompositie (/15)

Gegeven een vierkante  $n\times n$ matrix $\mathbf{A}$. De QR-decompositie van $\mathbf{A}$ wordt berekend door het uitvoeren van $n-1$ Householder reflecties.

In het bijzonder werkt de QR-decompositie als volgt:
- Stel `R` gelijk aan de originele matrix `A` (maar let op dat je `A` niet aanpast).
- Stel `Q` gelijk aan de eenheidsmatrix van grootte `n`.
- Voor `i` van `0` t.e.m. `n-2`:
    - Stel `k` gelijk aan de `i`-de kolom van `R`, maar stel de componenten met index kleiner dan `i` gelijk aan nul. (Gebruik hiervoor de methode `maak_nul`.)
    - Stel $u \gets k - \lVert k\rVert \mathbf{e}_i$, waarbij $\mathbf{e}_i$ de $i$-de eenheidsvector is. (Gebruik hiervoor de methode `eenheidsvector`.)
    - Stel `H` gelijk aan de matrix die een Householder reflectie voorstelt met als argument `u`. (Gebruik hiervoor de methode `maak_spiegeling`.)
    - Stel `R` gelijk aan de matrixproduct van `H` en `R`.
    - Stel `Q` gelijk aan de matrixproduct van `Q` en de getransponeerde van `H`.
  
Retourneer een tupel `(Q, R)`.

Belangrijke opmerking: na het uitvoeren van de methode mag de matrix `A` niet aangepast zijn.

````python
def QR(A):
    R = A.copy()
    n = len(A)
    Q = np.eye(n)
    # for i in range(n-2):
    for i in range(n):
        k = R[:,i]
        k = maak_nul(k, i)
        u = k - np.linalg.norm(k) * eenheidsvector(len(k), i)
        H = maak_spiegeling(u)
        R = H @ R
        Q = Q @ H.T
    return (Q, R)
````

````python
A = np.arange(16).reshape(4,4)
(Q, R) = QR(A)
````

- of de matrix $\mathbf{Q}$ orthogonaal is.

````python
for _ in range(100):
    rng = np.random.default_rng()
    A = rng.standard_normal((10,10))
    (Q, R) = QR(A)
    assert np.allclose(A, Q @ R), "A is niet gelijk aan Q * R"
    assert np.allclose(Q.T @ Q, np.eye(10)), "Q is niet orthogonaal"
    assert np.allclose(R, np.triu(R)), "R is niet bovendriehoek"
````
