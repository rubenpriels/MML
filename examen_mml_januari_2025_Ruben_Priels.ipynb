#### Methode `maak_nul` (/2)

Implementeer een methode `maak_nul(x, index)` die een numpy array `x` en een index `index` meekrijgt en een array retourneert waarbij alle elementen vanaf index `0` t.e.m. index `index-1` gelijk zijn aan nul. Alle andere elementen blijven ongewijzigd.  Je mag de originele array `x` niet aanpassen! Hieronder een voorbeeld:

```python
maak_nul(np.array([-1, -2, 3, 4, 5]), 2)  # array([0,0,3,4,5])
```

````python
def maak_nul(x, index):
    ### JOUW CODE HIER ###
    a = np.copy(x)
    for i in range(index) :
      a[i] = 0
    return a
    ### EINDE JOUW CODE HIER ###
````
